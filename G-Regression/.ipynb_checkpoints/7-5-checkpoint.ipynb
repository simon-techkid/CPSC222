```python
# Import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
```

```python
# Load the Wisconsin Breast Cancer dataset
PATH = "/Users/depalma/Desktop/CPSC222/A-Data/"
fin = PATH + "WisconsinBreastCancerDatabase.csv"
WBCD = pd.read_csv(fin)
# Convert Diagnosis to 0 and 1.
WBCD.loc[WBCD['Diagnosis'] == 'B', 'Diagnosis'] = 0
WBCD.loc[WBCD['Diagnosis'] == 'M', 'Diagnosis'] = 1
WBCD
```

```python
# Store relevant columns as variables
# Notice the transformation to 2-D
#values converts to a numpy array
# reshape: yo figure out how many rows.  Make it a single column
X = WBCD[['Radius mean']].values.reshape(-1, 1) 
y = WBCD[['Diagnosis']].values.reshape(-1, 1).astype(int)
```

```python
# Logistic regression predicting diagnosis from tumor radius
logisticModel = LogisticRegression()

# Compute b0 and b1 using maximum likelihood estimation:  find the parameters that make the observed data most plausible under the
# statistical model, in this case logistic regressio.  
logisticModel.fit(X, np.ravel(y.astype(int)))

# Plot Y against X, giving two horizontal lines.  Tumors are malignant or they are not
plt.scatter(X, y) 


#create 10,000 evenly spaced x values
xDelta = np.linspace(X.min(), X.max(), 10000)

'''
Evaluate probabilities across a smooth, simulated continuum of radius values.
xDelta is not the real data — it’s 10 000 evenly spaced radius values between the smallest and largest observed radius in the dataset.
For each simulated radius value, the model computes the probability of malignancy, using b0 and b1 as computed above and he 
the logistic regression function.  This produces a smooth S-shaped curve that visually represents the logistic function. I 
hope you see that it;s a a conditional probability: p(breast cancer | tumor radius)
'''
yDeltaProb = logisticModel.predict_proba(xDelta.reshape(-1, 1))[:, 1]

plt.plot(xDelta, yDeltaProb, color='red')
plt.xlabel('Radius', fontsize=14);
plt.ylabel('Probability of malignant tumor', fontsize=14);
```

```python
'''
X here is the real dataset: the measured mean radius for each tumor in the Wisconsin dataset.
The model looks at each actual radius and decides whether it predicts benign (0) or malignant (1), using a threshold of 0.5 on the logistic curve.
The result is a sequence of discrete predictions aligned with your real samples.
'''
#compute the predicted value, given X.  This produces the two horizontal lines.It is discrete rather than a probability. 
yPredicted = logisticModel.predict(X).reshape(-1, 1).astype(int)
plt.scatter(X, yPredicted) 
```

```python
# Display the slope parameter estimate, b1, as computed above
logisticModel.coef_
```

```python
# Display the intercept parameter estimate, b0 as computed above
logisticModel.intercept_
```

```python
# Predict the probability a tumor with radius mean 13 is benign / malignant
pHatProb = logisticModel.predict_proba([[13]])
#returns the probability that tumor is benign (col 0) or malignant (col 1)
#See below
pHatProb[0]
```
