```python
import matplotlib.pyplot as plt
```

```python
def plot_zipf(x,y):
    pl = plt.gca()
    pl.plot(x,y)
    pl.set_xlabel("Rank")
    pl.set_ylabel("Probability")
    #pl.plot("b",x,y)
    plt.show()
```

```python
def make_data(limit):
    x = [i for i in range (1,limit)]
    y = [1/i for i in x]
    return x,y
        
```

```python
x,y = make_data(500)
plot_zipf(x,y)
```
