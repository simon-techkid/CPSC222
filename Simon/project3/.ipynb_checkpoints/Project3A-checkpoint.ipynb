```python
from typing import List, Dict
import re
```

```python
MM: str = "middlemarch.txt"
UL: str = "ulysses.txt"
MR: str = "the_masque_of_the_red_death.txt"
```

```python
files: Dict[str, str] = {
    "MM": MM,
    "UL": UL,
    "MR": MR
}
```

```python
FILE: str = files["MM"]
```

```python
def open_file(filePath: str):
    while (True):
        try:
            fin = open(filePath, 'r')
            break
        except:
            print("Invalid file name, Try again")
    return fin
```

```python
def tokenize(string_in: str) -> List[str]:
    #substitute space for eol character
    string: str = re.sub('\n',' ', string_in)

    #create a list containing all lower case characters
    good_chars: List[str] = [chr(value) for value in range(ord('a'),ord('z') + 1,1)]
    good_chars.append(' ') #spaces are good characters
    string = string.lower()

    new_str: str = ''
    for ch in string:
        if ch in good_chars:
            new_str = new_str + ch
    
    #remove spaces
    lst: List[str] = new_str.split()
    return lst 
```

```python
fin = open_file(FILE)
fst: str = fin.read()
lst: List[str] = tokenize(fst)
```

```python
dic: Dict[str, int] = {}
for word in lst:
    if word in dic:
        dic[word] = dic[word] + 1
    else:
        dic[word] = 1

```
