```python
#Pandas Aggregation Functions
count = int(country["Country"].count())
print(count)
mean = float(country["Population"].mean())
median = int(country["Density"].median())
min = country["Population"].min() 
print(min)
#also std, var, prod, sum for columns in data frames
```

```python
#Next several cells will illustrate Pandas GroupBy functions
#Group by splits dataframe into subsets based on values in columns--here A, B, C

df = pd.DataFrame({'key': ['A', 'B', 'C', 'A', 'B', 'C'],
                   'data': range(6)}, columns=['key', 'data'])
df

print((df.key,df.data))
```

```python
#Returns a group by object,just waiting for an aggregation function to be applied
#Group by splits dataframe into subsets based on values in columns--here A, B, C

x = df.groupby('key')
for item in x:
    print(item)
```

```python
df.groupby('key').sum()   #i.e, sum over A, B, C
```

```python
for item in df.groupby('key'):
    print(item)  #there are two rows with each key
```

```python
df.groupby('key').aggregate(['min', 'median', 'max'])
```

```python
#Another example, slightly more complex

rng = np.random.RandomState(0)  #set seed value
df = pd.DataFrame({'key': ['A', 'B', 'C', 'A', 'B', 'C'],
                   'data1': range(6),
                   'data2': rng.randint(0, 10, 6)},       #produce 6 randome integers in [0..10]
                  
                   columns = ['key', 'data1', 'data2'])
df
```

```python
#Now lets aggregate: reduce each group to a anmuber

df.groupby('key').aggregate(['min', 'median', 'max'])
```

```python
#Pivot Table extracts from an original table a column with respect to  other columens
## Ex. What is the average years of education of each continent
#It produces exactly what group by produces but in a dataframe which can then be manipulated.  

pt = country_comp.pivot_table(
    values = 'Years', index = 'Continent', aggfunc='mean')
pt
```

```python
#This can be more complex.  Suppose we want by aggregate by fertility, too
pt1 = country_comp.pivot_table(
    values = ['Years','Fertility'], index = 'Continent', aggfunc='mean')
pt1
```
