```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
PATH = "/Users/depalma/Desktop/CPSC222/A-Data/"
DS1 = "country.csv"
DS2 = "country_complete.csv"
fin1 = PATH + DS1
fin2 = PATH + DS2

country = pd.read_csv(fin1)     #specifies the column headers
country_comp = pd.read_csv(fin2)
print(country)
#country.info()  #returns data types for columns
```

```python
#Pandas Aggregation Functions
count = int(country["Country"].count())
mean = float(country["Population"].mean())
median = int(country["Density"].median())
min = country["Population"].min() 
#also std, var, prod, sum for columns in data frames
```

```python
#Next several cells will illustrate Pandas GroupBy functions
#Group by splits dataframe into subsets based on values in columns--here A, B, C

df = pd.DataFrame({'key': ['A', 'B', 'C', 'A', 'B', 'C'],
                   'data': range(6)}, columns=['key', 'data'])
df
```

```python
#Returns a group by object,just waiting for an aggregation function to be applied
#Group by splits dataframe into subsets based on values in columns--here A, B, C

df.groupby('key')
```

```python
df.groupby('key').sum()   #i.e, sum over A, B, C
```

```python
for item in df.groupby('key'):
    print(item)  #there are two rows with each key
```

```python
df.groupby('key').aggregate(['min', 'median', 'max'])
```

```python
#Another example, slightly more complex

rng = np.random.RandomState(0)  #set seed value
df = pd.DataFrame({'key': ['A', 'B', 'C', 'A', 'B', 'C'],
                   'data1': range(6),
                   'data2': rng.randint(0, 10, 6)},       #produce 6 randome integers in [0..10]
                  
                   columns = ['key', 'data1', 'data2'])
df
```

```python
#Now lets aggregate: reduce each group to a anmuber

df.groupby('key').aggregate(['min', 'median', 'max'])
```

```python
#Pivot Table extracts from an original table a column with respect to  other columens
## Ex. What is the average years of education of each continent
#It produces exactly what group by produces but in a dataframe which can then be manipulated.  

pt = country_comp.pivot_table(
    values = 'Years', index = 'Continent', aggfunc='mean')
```

```python
print(pt)
float(pt['Years'].max())

```

```python
#This can be more complex.  Suppose we want aggregate by fertility, too
pt1 = country_comp.pivot_table(
    values = ['Years','Fertility'], index = 'Continent', aggfunc='mean')
pt1
```

```python
# cat.reorder_categories is useful for rearranging the order
# (ex: low to high)
country['Internet access'] = country['Internet access'].cat.reorder_categories(
    ['Low', 'Moderate', 'High', 'Very high']
)
# Display the number of countries in a pivot table of continent and
# internet access
country.pivot_table(
    values='Years', index='Continent', columns='Internet access', aggfunc=np.size
)
```

```python
# Which 7 countries in the Americas have low Internet access?
country[(country['Continent'] == 'Americas') & (country['Internet access'] == 'Low')]
```

```python
# Display the average number of years of schooling
country.pivot_table(
    values='Years', index='Continent', columns='Years', aggfunc=np.mean
)
```

```python
# Display the average of each feature by continent

country.groupby(by=["Continent"]).mean()
```

```python
# Display the median of each feature by continent

country.groupby(by=["Continent"]).median()
```

```python
# Display the number of countries in the dataset
# for each continent

country.groupby(by=["Continent"]).size()
```

```python
# Use describe on each feature for each continent

country.groupby(by=["Continent"]).describe()
```

```python
# Display a subset of the data above using .agg()

country.groupby("Continent").agg(
    min_Emissions=pd.NamedAgg(column="Emissions", aggfunc="min"),
    max_Emissions=pd.NamedAgg(column="Emissions", aggfunc="max"),
    mean_Emissions=pd.NamedAgg(column="Emissions", aggfunc=np.mean),
)
```
