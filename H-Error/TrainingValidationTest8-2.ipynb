```python
# Import packages
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
# Load bad drivers data
PATH = "/Users/depalma/Desktop/"
DSDATA = "CPSC222/A-Data/bad-drivers.csv"
badDrivers = pd.read_csv(PATH + DSDATA)
#print(badDrivers.head())
```

```python
# Display the bad drivers data
#badDrivers
```

```python
# Set the proportions of the training-validation-test split
trainingProportion = 0.70
validationProportion = 0.10
testProportion = 0.20

'''
The workings of train_test_split depend upon whether test_size or train_size is used as the second parameter
If test_size is used, test_size fraction is in the second return variable. The leftover is in the first
if train_size is used, train_size  fraction is in the first return variable. The leftover is in the second
'''


# test_size: TestData (2nd( is 20%, leftover, trainingAndValidationData (1st), is 80%
trainingAndValidationData, testData = train_test_split(
    badDrivers, test_size=testProportion
)

# train_size: train_fraction is in the 1st return variable, the leftover (validation data) is in the second,
#But these are both fractions of what's leftover after the first split.   So, the training_size, is the training size after the
#test data has been split off, namely: trainingProportion / (trainingProportion + validationProportion.  The test size is what's left, 
#the validation data
trainingData, validationData = train_test_split(
    trainingAndValidationData,
    train_size=trainingProportion / (trainingProportion + validationProportion)
)
```
