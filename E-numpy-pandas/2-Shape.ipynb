## Part 2: Shape Functions

```python
#
import numpy as np
#Arrays can be reshaped
array1 = np.array([[1, 2], [3, 4]])
print(array1)
print()
array1 = array1.reshape(1, 4)
print(array1)
array1 = array1.reshape(2,2)
print(array1)

```

```python
print(array1.shape)
#we can flatten this weird array
array1 = np.ravel(array1)
print(array1)
```

```python
array1 = np.array([[1, 2], [3, 4]])
print(array1)
print()
array1 = np.reshape(array1, (1, 4))
print(array1)

print(array1.shape)

array1 = np.reshape(array1,(2,2))
print(array1)
```

```python
#make array1 2x2

array1 = np.array([[2, 4], [6, 8]])
print(array1)
array1 = array1.reshape(1,4)
print(array1)
array1 = np.ravel(array1)
print(array1)


```

```python
#transpose the rows and columns
array1 = array1.reshape(2,2)
array1 = np.transpose(array1)
print(array1)
```
