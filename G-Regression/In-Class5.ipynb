```python
#Plot age against the prbability of having diabetes in the nhanes dataset
data = PATH + "nhanes.csv"
nhanes = pd.read_csv(data)
#print(nhanes.columns)
```

```python
#1 Make diabetes numeric, because we'll count it. Add a new column
nhanes = nhanes.dropna(subset=["Diabetes"])
nhanes["Diabetes_num"] = nhanes["Diabetes"].map({"Yes": 1, "No": 0})

#Group by age.  Compute the mean over each age
age_diabetes = nhanes.groupby("Age")["Diabetes_num"].mean().reset_index()


```

```python
plt.plot(age_diabetes["Age"], age_diabetes["Diabetes_num"], marker="o")
plt.xlabel("Age")
plt.ylabel("Proportion with Diabetes")
plt.title("Probability of Diabetes by Age")
plt.show()
```
