```python
import numpy as np
import pandas as pd
from pandas.api.types import is_numeric_dtype
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsRegressor
from sklearn import preprocessing
```

```python
PATH = "/Users/depalma/Desktop/TA/CPSC222/A-Data/BostonHousing/"
fin = "BostonHousing.csv"
fin = PATH + fin
BH = pd.read_csv(fin)
```

```python
#Clean Data
print(BH.shape)
BH = BH.dropna()
print(BH.shape)

#Make sure relevant columns are numreric
from pandas.api.types import is_numeric_dtype

if is_numeric_dtype(BH['RM']):
    print('ok')
if is_numeric_dtype(BH['LSTAT']):
    print('ok')


#Normalize Date
#See feature scaling, zybooks 5.3.2 Formatting Data
#print(BH.head())

#Normalize only relevant columns
cols_to_scale = ['RM', 'LSTAT']
normalizer = preprocessing.MinMaxScaler()
normArray = normalizer.fit_transform(BH[cols_to_scale])

#Replace those columns with the normalized values from the normalizer
BH[cols_to_scale] = normArray
#print()
#print(BH.head())
```

```python
# Fit KNN model with k = 5
# Store relevant columns as variables
X = BH[['RM', 'LSTAT']].values.reshape(-1, 2)
#X = BH[['RM', 'LSTAT']]
y = BH[['MEDV']].values.reshape(-1, 1)
knn = KNeighborsRegressor(n_neighbors=5)
knn.fit(X, y);
```

```python
test_point = X[100]  # arbitrary example
predicted_price = knn.predict([test_point])[0]
print(predicted_price);
```
