```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split
#from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
PATH = "/Users/depalma/Desktop/TA/CPSC222/A-Data/BostonHousing/"
fin = "BostonHousing.csv"
fin = PATH + fin
BH = pd.read_csv(fin)
```

```python
#Clean Data
#print(BH.shape)
BH = BH.dropna()
#print(BH.shape)

#Make sure relevant columns are numreric
from pandas.api.types import is_numeric_dtype

#if is_numeric_dtype(BH['RM']):
#    print('ok')
#if is_numeric_dtype(BH['LSTAT']):
#    print('ok')

```

```python
#Create an illness vulnerability column in BH
lstat_p70 = np.percentile(BH['LSTAT'], 70)  # value of LSTAT at 70th percentile (70% concentration of low income)
medv_p30 = np.percentile(BH['MEDV'], 30)    # bottom 30% MEDV (value of MEDV at 30th percentile (i.e., lowest 30%)

BH['Vulnerable'] = (
    (BH['LSTAT'] >= lstat_p70) &
    (BH['MEDV'] <= medv_p30)
).astype(int)
#print(BH.head)
```

```python
#Prelims
# Store relevant columns as variables
X = BH[['LSTAT','MEDV', 'RM']].values.reshape(-1, 3) #extract two cols, convert to numpy array with 3 features per row
y = BH[['Vulnerable']].values.reshape(-1, 1) #as above but ensure that y is a column vector
#print(X[:5])   # First 5 samples of features
#print(y[:5])   # First 5 target values

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)  #stratify ensures that the train/test split gets the same proportion of records that fall into Vulnerable/not vulnerable
y_train = y_train.ravel()
y_test = y_test.ravel()

#Normalize/scale
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train) #find min/max params (fit) then tranform the data 

X_test_scaled = scaler.transform(X_test) #mins/maxes found from training set are available to the scaler. Don't find new ones from
#the test data.
```

```python
#Fit Data
k = 5  # you can tune this
knn = KNeighborsClassifier(n_neighbors=k)
knn.fit(X_train_scaled, y_train)
#print(X_train.shape)
#print(y_train.shape)
#print(X_test.shape)
#print(y_test.shape)

#y_pred = knn.predict(X_test_scaled)

#print(f"\nAccuracy (k={k}): {accuracy_score(y_test, y_pred):.3f}")
#print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
BH = BH.reset_index(drop=True)  #necessary because rows were dropped during data cleaing

idx = 325
test_point = BH.loc[idx, ['LSTAT', 'MEDV','RM']].values.reshape(1, -1)

# Scale using the same scaler as the training data
#test_point_scaled = scaler.transform(test_point)

predicted_class = knn.predict(test_point)[0]

# Get coordinates for plotting
x_val = BH.loc[idx, 'LSTAT']
y_val = BH.loc[idx, 'MEDV']
```

```python
plt.figure(figsize=(7, 5))
scatter = plt.scatter(
    BH['LSTAT'],
    BH['MEDV'],
    c=BH['Vulnerable'],
    cmap='coolwarm',
    alpha=0.7,
    edgecolor='k',
    linewidth=0.3
)

# Highlight selected data point in GREEN
plt.scatter(
    x_val, y_val,
    marker='o',
    s=200,
    edgecolor='black',
    facecolor='green',    # <-- GREEN color requested
    linewidth=2,
    label=f"Point {idx} (pred={predicted_class})"
)


plt.xlabel("LSTAT (% lower-status population)")
plt.ylabel("MEDV (median home value, $1000s)")
plt.title("Neighborhood Vulnerability")

handles, _ = scatter.legend_elements()  #extracts color items from matplotlib to create legend bosd
labels = ["Lower vulnerability (0)", "Higher vulnerability (1)"]
plt.legend(handles, labels, title="Vulnerability Proxy")

#plt.gca().invert_yaxis()  
plt.tight_layout()
plt.show()
```
