# Descriptive Statistics: Populations and Samples

```python
import pandas as pd
import matplotlib.pyplot as plt  # used to create a histogram
```

```python
PATH = "/Users/depalma/Desktop/CPSC222/A-Data/"
data = PATH + "nhanes.csv"
nhanes = pd.read_csv(data)
```

```python
#Take 5 random samples of 20 records witout replacement
#samp_lst = [nhanes.sample(n=20, replace=False) for i in range(5)]
sample_lst1 = [nhanes.sample(n=20, replace=False) for i in range(5)]
```

```python
#Compute the sample means
sample_means1 = [our_sample["Age"].mean() for our_sample in sample_lst1]
#Compute the sampe standard deviations
sample_std1 = [our_sample["Age"].std() for our_sample in sample_lst1]
```

```python
for i in sample_means1:
    print(i)
print()
for i in sample_std1:
    print(i)
```

```python

sample_lst2 = [nhanes.sample(n=200, replace=False) for i in range(5)]
means2 = [our_sample["Age"].mean() for our_sample in sample_lst2]
#Compute the sampe standard deviations
std2 = [our_sample["Age"].std() for our_sample in sample_lst2]

for i in means2:
    print(i)
print()
for i in std2:
    print(i)
```

```python
#compute the population mean
pop_mean = nhanes["Age"].mean()
print(pop_mean)
```

```python
#expanding the number of bins: each covers a smaller range of values.  
plt.hist(x=nhanes["Age"],bins=25)
plt.show()
```

```python
#very little data.  Histogram doesn't tell much
plt.hist(x=sample_lst1[0]["Age"],bins=5)
plt.show()
```
